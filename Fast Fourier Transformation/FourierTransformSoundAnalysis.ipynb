```python
import numpy as np
import numpy.fft as ffft
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
fs = 1000
t = np.arange(0,1.5,1/fs)
tone1 = 200
tone2 = 300
tone3 = 400
sinyal = np.cos(2*np.pi*tone1*t)+1/2*np.cos(2*np.pi*tone2*t)+1/3*np.cos(2*np.pi*tone3*t)
s = ffft.fft(sinyal,fs)
s_abs = np.absolute(s)

print('t = ', t)
print('sinyal = ', sinyal)
print('s = ', s)
print('s_abs = ', s_abs)
```

```python
mpl.rcParams['figure.dpi'] = 300
plt.figure()
fig, ax = plt.subplots(figsize=(10,10))
plt.tight_layout
plt.subplot(3,1,1); plt.plot(t,sinyal); plt.axis([0, 0.2, -1.5, 1.5]); plt.xlabel('Waktu (S)'); plt.ylabel('Amplitudo')
plt.subplot(3,1,2); plt.plot(s_abs); plt.xlabel('Frekuensi (Hz)'); plt.ylabel('Magnitudo')
plt.savefig('fouriertransformsoundanalisis.png', bbox_inches='tight')
plt.show()
```
