```python
import numpy as np
import numpy.fft as np_fft
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#singal generator
fs = 200
dt = 1/fs
t = np.arange(0,0.1,dt)
x1 = np.cos((2*np.pi*t)/3)
x2 = np.sin((2*np.pi*t)/5)
x = x1+x2
```

```python
#Fourier Transform
y = np_fft.fft(x)
f = 200*np.arange(0,len(x1))/len(x1)
z = abs(y)

print('f = ', f)
print('z = ', z)
```

```python
mpl.rcParams['figure.dpi'] = 300
plt.figure()
fig, ax = plt.subplots(figsize=(10,10))
plt.tight_layout
plt.subplot(2,2,1); plt.plot(t,x1); plt.grid(); plt.title('Sinyal Input Cosinus'      ); plt.xlabel('Waktu'    ); plt.ylabel('Amplitudo')
plt.subplot(2,2,2); plt.plot(t,x2); plt.grid(); plt.title('Sinyal Input Sinus'        ); plt.xlabel('Waktu'    ); plt.ylabel('Amplitudo')
plt.subplot(2,2,3); plt.plot(t,x ); plt.grid(); plt.title('Sinyal Input'              ); plt.xlabel('Waktu'    ); plt.ylabel('Amplitudo')
plt.subplot(2,2,4); plt.plot(f,z ); plt.grid(); plt.title('Hasil Transformasi Fourier'); plt.xlabel('Frekuensi'); plt.ylabel('Magnitudo')
plt.savefig('fouriertransformtestsin.png', bbox_inches='tight')
plt.show()
```
