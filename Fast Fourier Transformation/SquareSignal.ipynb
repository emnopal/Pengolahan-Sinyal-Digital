```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
f = 1
Fs = 2 
n = 5
t = np.linspace(0,3,301)
sq = np.zeros(len(t)) 
h = np.arange(1,n,2) 
sqwave = np.sign(np.sin(2*np.pi*f*t))
for i in h:
    sq += (4/(np.pi*i))*np.sin(2*np.pi*f*i*t)

mpl.rcParams['figure.dpi'] = 300
plt.axis([0, 3, -1.25, 1.25])
plt.plot(t,sq)
plt.savefig('squaresignal5.png', bbox_inches='tight')
plt.show()

print('sq = ', sq)
print('t = ', t)
print('h = ', h)
print('sqwave = ', sqwave)
```
