```python
import numpy as np
import numpy.fft as np_fft
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#singal generator
fs = 500
f1 = 20
f2 = 50
dt = 1/fs
t = np.arange(0,0.1,dt)
x1 = np.cos(2*np.pi*f1*t)
x2 = np.cos(2*np.pi*f2*t)
```

```python
#Fourier Transform
y1 = np_fft.fft(x1)
y2 = np_fft.fft(x2)
f = 500*np.arange(0,len(x1))/len(x1)
z1 = abs(y1)
z2 = abs(y2)

print('f = ', f)
print('z1 = ', z1)
print('z2 = ', z2)
```

```python
mpl.rcParams['figure.dpi'] = 300
plt.figure()
fig, ax = plt.subplots(figsize=(10,10))
plt.tight_layout
plt.subplot(2,2,1); plt.plot(t,x1); plt.grid(); plt.title('Sinyal Input'); plt.xlabel('Waktu'); plt.ylabel('Amplitudo')
plt.subplot(2,2,2); plt.plot(f, z1); plt.grid(); plt.axis([0, 250, 0, 30]); plt.title('Hasil Transformasi Fourier'); plt.xlabel('Frekuensi'); plt.ylabel('Magnitudo')
plt.subplot(2,2,3); plt.plot(t,x2); plt.grid(); plt.title('Sinyal Input'); plt.xlabel('Waktu'); plt.ylabel('Amplitudo')
plt.subplot(2,2,4); plt.plot(f, z2); plt.grid(); plt.axis([0, 250, 0, 30]); plt.title('Hasil Transformasi Fourier'); plt.xlabel('Frekuensi'); plt.ylabel('Magnitudo')
plt.savefig('fouriertransformtestcos.png', bbox_inches='tight')
plt.show()
```
